```python
import pandas as pd
from collections import defaultdict
```

```python
df_par = pd.read_csv('../../static/parkinson_1960tonow.csv', index_col=0)
df_par.date = pd.to_datetime(df_par.date)
df_par = df_par[(df_par.date.dt.year >= 1999) & (df_par.date.dt.year <= 2017)].sort_values(by='date')
df_par.head()
```

```python
jr_dict = defaultdict(list)
for year in range(1999,2018):
    scimago = pd.read_csv('../../static/scimago/scimagojr {}  \
Subject Category - Neurology (clinical).csv'.format(year), sep=';')
    jr_dict[year] = scimago.Title[:20].str.lower().tolist()
```

```python
top_counts = {}
for year in range(1999, 2018):
    cnt = df_par[df_par.date.dt.year == year]['published in'].apply(lambda x: x.lower() in jr_dict[year]).sum()
    top_counts[year] = cnt
```

```python
top_counts
```

```python
top_jrs = []
for year in range(1999, 2018):
    journals = []
    mask = df_par[df_par.date.dt.year == year]['published in'].apply(lambda x: x.lower() in jr_dict[year]) 
    for x in df_par[df_par.date.dt.year == year]['published in'].loc[mask].values:
        top_jrs.append((year, x))
```

```python
top_jrs
```

```python
df_topjr = pd.DataFrame(top_jrs, columns = ['year', 'top_journal'])
df_topjr.to_csv('../../static/dash_data/top_jr.csv')
```

```python
df_topjr
```

```python
from collections import Counter
```

```python
topjr_hover = []
for i in range(1999, 2017):
    topjr_hover.append((i, str(Counter(df_topjr[df_topjr.year==i].top_journal.values))[9:-2]))
```

```python
df_topjr_hover = pd.DataFrame(topjr_hover, columns = ['year', 'journal_counts'])
df_topjr_hover.journal_counts=df_topjr_hover.journal_counts.apply(lambda x: x.replace(', ', '<br>'))
```

```python
df_topjr_hover.to_csv('../../static/dash_data/topjr_hover.csv')
```

```python
import numpy as np
df_counts = pd.DataFrame(df_par.groupby(df_par.date.dt.year).title.count())
df_counts.columns = ['total_count']
df_counts['top_count'] = top_counts.values()
df_counts['top_fraction'] = df_counts.top_count / df_counts.total_count
df_counts['top_percentage'] = np.round_(df_counts.top_fraction * 100, 2)
df_counts
```

```python
df_counts.drop('top_fraction', axis=1).to_csv('../../static/dash_data/top_jr_counts.csv', index_label='date')
```
